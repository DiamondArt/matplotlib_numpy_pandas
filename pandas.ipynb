```python
#verifier s'il y a des donnees manquantes
df.isnull().sum()
```

```python
#suppression les lignes ou il y a des donnees maquantes
df.dropna(how='all', inplace=True)
```

```python
#Recodage de sexe (0 pour masculin et 1 pour féminin)
df['sexe'] = df['sexe'].replace(['masculin','feminin'],['0','1'])
df
```

```python
#supprimer les variable depression, vaisseau et pic du dataframe
df.drop(['depression','vaisseau','pic'], axis=1)
```

```python
#Representation graphique de la variable coeur
fig = plt.figure(figsize = (10, 10))
plt.hist(df.coeur, edgecolor='white')
```

```python
#Representation graphique de toutes les variables qualitatives

var_qual= df.select_dtypes('object')
colors = ["blue","red","orange","grey","cyan","purple","black","green"]

for index, libelle in enumerate(var_qual):
    
    fig = plt.figure(figsize = (5, 5))
    plt.hist(df[libelle],color= colors[index], edgecolor = 'white', label=libelle)
    plt.legend()
    plt.title(libelle+" variable")
plt.show()
```

```python
#Representation graphique de toutes les variables quantitative

var_quant= df.select_dtypes('int')
for index, libelle in enumerate(var_quant):
    
    fig = plt.figure(figsize = (5, 5))
    plt.hist(df[libelle],color= colors[index], edgecolor = 'white', label=libelle)
    plt.legend()
    plt.title(libelle)
plt.show()
```
